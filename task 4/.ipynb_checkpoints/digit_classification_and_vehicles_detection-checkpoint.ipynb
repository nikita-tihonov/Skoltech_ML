```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import struct
import cv2
import torch

from glob import glob

from skimage.feature import hog
from skimage import data, exposure

from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC, LinearSVC
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report, ConfusionMatrixDisplay
```

# Useful function

```python
def load_idx(filename):
    with open(filename, 'rb') as f:
        zero, data_type, dims = struct.unpack('>HBB', f.read(4))
        shape = tuple(struct.unpack('>I', f.read(4))[0] for d in range(dims))
        return np.frombuffer(f.read(), dtype=np.uint8).reshape(shape)
```

```python
def print_metrics(y_test, y_pred, model):
    print("Accuracy: %f" % accuracy_score(y_test, y_pred))
    print(classification_report(y_test, y_pred, target_names=target_names))
    
    
    M = confusion_matrix(y_test, y_pred, labels=model.classes_)
    M_normalized = M.astype('float') / M.sum(axis=1)[:, np.newaxis]
    
    
    disp = ConfusionMatrixDisplay(confusion_matrix=M,
                                  display_labels=target_names)
    disp.plot(cmap='Greens')
    plt.xticks(ticks = range(len(target_names)), labels = target_names, rotation=90)
    plt.yticks(ticks = range(len(target_names)), labels = target_names)
    
    plt.ylabel('True digit')
    plt.xlabel('Predicted digit')
    plt.title('Confusion matrix')
    
    plt.show()
```

# Классификация рукописных символов

```python
train_images_path = 'datasets/train-images.idx3-ubyte'
train_labels_path = 'datasets/train-labels.idx1-ubyte'
test_images_path = 'datasets/t10k-images.idx3-ubyte'
test_labels_path = 'datasets/t10k-labels.idx1-ubyte'

X_train = load_idx(train_images_path)
y_train = load_idx(train_labels_path)
X_test = load_idx(test_images_path)
y_test = load_idx(test_labels_path)
```

```python
X_train.shape, y_train.shape, X_test.shape, y_test.shape
```

```python
# Смотрим датасет

rows = 4
cols = 8
fig, ax = plt.subplots(rows, cols, figsize=(15, 10))

num = 0
for i in range(rows):
    for j in range(cols):
        ax[i][j].imshow(X_train[1000 * num], cmap = 'gray')
        ax[i][j].set_title(y_train[1000 * num])
        ax[i][j].set_xticks([])
        ax[i][j].set_yticks([])
        num += 1
```

```python
# Распрямляем массивы 

X_train = [image.flatten() for image in X_train]
X_test = [image.flatten() for image in X_test]

X_train = np.array(X_train)
y_train = np.array(y_train)
X_test = np.array(X_test)
y_test = np.array(y_test)
target_names=list(map(str, np.unique(y_test)))
```

```python
pca = PCA(n_components=2)
X_reduced = pca.fit_transform(X_train)

print('Projecting %d-dimensional data to 2D' % np.shape(X_train)[1])

plt.figure(figsize=(12,10))
plt.scatter(X_reduced[:, 0], X_reduced[:, 1], c=y_train, 
            edgecolor='none', alpha=0.3, s=40,
            cmap=plt.cm.get_cmap('nipy_spectral', 10))
plt.colorbar()
plt.show()
```

```python
# Смотрим сколько информации остается при соответсвующем количестве компонент

pca = PCA().fit(X_train)

plt.figure(figsize=(10,7))
plt.plot(np.cumsum(pca.explained_variance_ratio_), color='k', lw=2)
plt.xlabel('Number of components')
plt.ylabel('Total explained variance')
plt.yticks(np.arange(0, 1.1, 0.1))
plt.axvline(np.where(np.cumsum(pca.explained_variance_ratio_) > 0.9)[0][0], c='b')
plt.axhline(0.9, c='r')
plt.show()
```

```python
# Оставляем 100 компонент

pca = PCA(n_components=100).fit(X_train)
```

```python
X_train_pca = pca.transform(X_train)
X_test_pca = pca.transform(X_test)
```

```python
# Выбираем подвыборку из 1000 случайных элементов датасета, чтобы найти наилучшие гиперпараметры

mask = np.random.choice(list(range(0, len(X_train_pca))), size=1000)

param_grid = [
  {'C': [1, 5, 10], 'kernel': ['linear']},
  {'C': [1, 5, 10], 'gamma': ['scale'], 'kernel': ['rbf']},
  {'C': [1, 5, 10], 'degree': [1, 2, 3, 4, 5, 6, 7, 8, 9], 'gamma': ['scale'], 'kernel': ['poly']},
  {'C': [0.1, 1, 5, 10], 'kernel': ['sigmoid']}
]
svc = SVC()
clf = GridSearchCV(svc, param_grid)

clf.fit(X_train_pca[mask], y_train[mask])
```

```python
print('Наилучшие гиперпараметры: ', clf.best_params_)
```

```python
df_res = pd.DataFrame(clf.cv_results_)[['mean_fit_time', 'params', 'mean_test_score']]
df_res[df_res.mean_test_score > 0.85]
```

Видим, что с задачей хорошо справляются модели rbf (C = 5) и poly (degree = 3, C = 5)

Обучим модель rbf на всем датасете

```python
%%time
model = SVC(kernel='rbf', C=5)
model.fit(X_train_pca, y_train)
y_pred = model.predict(X_test_pca)
```

```python
print_metrics(y_test, y_pred, model)
```

```python
def get_heatmap(image, pca, model):
    step = 32
    size_of_sliding_window = 96
    upper_threshold = image.shape[0]//2
    heatmap_image = np.zeros(image[:, :, 0].shape)
    sliding_windows = np.lib.stride_tricks.sliding_window_view(image, (size_of_sliding_window, size_of_sliding_window, 3))[upper_threshold::step, ::step]
    for i in range(sliding_windows.shape[0]):
        for j in range(sliding_windows.shape[1]):
            slice = sliding_windows[i][j][0]
            slice = cv2.resize(slice, (64, 64))
            slice = cv2.cvtColor(slice, cv2.COLOR_BGR2YCrCb)
            fd, hog_image = hog(slice, orientations=9, pixels_per_cell=(8, 8),
                    cells_per_block=(2, 2), visualize=True, channel_axis=-1)
            H = slice[:, :, 0].copy().flatten()
            S = slice[:, :, 1].copy().flatten()
            V = slice[:, :, 2].copy().flatten()
            slice_flatten = np.concatenate((H, S, V, hog_image.flatten()))
            slice_flatten = slice_flatten.reshape((1, slice_flatten.shape[0]))
            slice_pca = pca.transform(slice_flatten)
            y_proba = model.predict_proba(slice_pca)
            if y_proba[0][1] > 0.9:
                heatmap_image[upper_threshold + step * i: upper_threshold + step * i + size_of_sliding_window:, step * j: step * j + size_of_sliding_window:] += 1
    return heatmap_image
```

```python
def plot_bbox(heatmap_image, image):
    heatmap_image = heatmap_image.astype(np.uint8)
    contours, hierarchy = cv2.findContours(heatmap_image, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)
    for cnt in contours:
        x, y, w, h = cv2.boundingRect(cnt)
        cv2.rectangle(image, (x, y), (x + w, y + h), (0, 255, 0), 2)
    return image
```

```python
%%time
heatmap_image = get_heatmap(image, pca, model)
```

## Видеопоток

```python
filename = './BDD_videos/LaGuardia_New_York.mov'
cap = cv2.VideoCapture(filename)

if (cap.isOpened()== False): 
  print("Error opening video stream or file")
```

```python
video = []
i = 0
while(cap.isOpened()):
    ret, frame = cap.read()
    if ret == True:
        i += 1
        if i % 20 == 0:
            heatmap_image = get_heatmap(frame, pca, model)
            frame = plot_bbox(heatmap_image, frame)
            
            cv2.imshow('Frame', frame)
    
        if cv2.waitKey(25) & 0xFF == ord('q'):
            break
    else: 
        break
 
cap.release()
cv2.destroyAllWindows()
```
